```python
#importing neccessory libraries
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from sklearn.svm import SVC
from sklearn.metrics import r2_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

#loading data 
df = pd.read_csv('leaf.csv')

#renaming the columns
df.columns = columns=['Class_Species','Specimen Number', 'Eccentricity','Aspect Ratio', 'Elongation', 'Solidity','Stochastic Convexity','Isoperimetric Factor','Maximal Indentation Depth','Lobedness', 'Average Intensity','Average Contrast', 'Smoothness','Third moment','Uniformity', 'Entropy']

print('Shape of dataset is :', df.shape)

# dividing input and output columns
X = df.drop('Class_Species', axis=1)
y = df['Class_Species']

# Splitting the dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
print('\n')
print('X_train shape: ', X_train.shape)
print('X_test shape: ', X_test.shape)
print('y_train shape: ', y_train.shape)
print('y_test shape: ', y_test.shape)

#declaring model
svc = SVC(C=1.0,
    kernel='poly',
    degree=30)

svc.fit(X_train, y_train)
print('\nAccuracy score of model is: ', svc.score(X_train, y_train))

train_pred = svc.predict(X_train)
print('\nr-squarred value for training set is: ',r2_score(y_train, train_pred))
print('Classification report of training set is: \n', classification_report(y_train, train_pred))

print('-'*100)

y_pred = svc.predict(X_test)
print('\nr-squarred value for testing set is: ', r2_score(y_test, y_pred))
print('Classification report of testing set is: \n',classification_report(y_pred, y_test))
```
